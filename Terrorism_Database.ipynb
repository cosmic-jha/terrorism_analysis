```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('terror_db.csv', low_memory = False)
```

```python
country_attacks = data['country_txt'].value_counts()
top_countries = country_attacks.head(10)
year_attacks = data['iyear'].value_counts()
top_years = year_attacks.head(10)
```

```python
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Maximum Terror Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_years.plot(kind='bar', color='red')
plt.title('Top 10 year with most attacks')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
year_counts = data['iyear'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(year_counts.index, year_counts.values, marker='o')
plt.title('Temporal Trends of Terrorism Incidents')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.show()
```

```python
attack_counts = data['attacktype1_txt'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(attack_counts.index, attack_counts.values)
plt.title('Distribution of Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()

```

```python
target_counts = data['targtype1_txt'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(target_counts.index, target_counts.values)
plt.title('Distribution of Target Types')
plt.xlabel('Target Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=90)
plt.show()
```

```python

gang_counts = data['gname'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gang_counts, labels=gang_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.axis('equal')
plt.title('Distribution of Gang Involvement')
plt.show()
```

```python
gang_counts = data['gname'].value_counts()
top_categories = gang_counts.head(15)
other_count = gang_counts[15:].sum()
top_categories['Others'] = other_count
plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.axis('equal') 
plt.title('Distribution of Gang Involvement (Top 15)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['country_txt'], data['weaptype1_txt'], color='blue')
plt.title('Weapons Used in Incidents')
plt.xlabel('Incident ID')
plt.ylabel('Number of Weapons Used')
plt.grid(True)
plt.show()
```

```python
top_weapons = data.groupby('country_txt')['weaptype1_txt'].head(15)
plt.figure(figsize=(10, 6))
plt.hist(top_weapons, bins=10, edgecolor='black', alpha=0.7)
plt.title('Top 15 Weapons Used per Country')
plt.xlabel('Number of Weapons Used')
plt.ylabel('Frequency')
plt.grid(True)
plt.xticks(rotation = 90)
plt.show()
```
